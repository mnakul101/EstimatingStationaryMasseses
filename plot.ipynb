```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
c0 = np.load("WingIT+PI_TV_error_n0=d_c=0.0.npy")
c01 = np.load("WingIT+PI_TV_error_n0=d_c=0.1.npy")
c02 = np.load("WingIT+PI_TV_error_n0=d_c=0.2.npy")
c03 = np.load("WingIT+PI_TV_error_n0=d_c=0.33.npy")
c04 = np.load("WingIT+PI_TV_error_n0=d_c=0.4.npy")
c05 = np.load("WingIT+PI_TV_error_n0=d_c=0.5.npy") 
```

```python
powers = np.linspace(6,10,10)

n0_list=[]
for power in powers:
    n0_list.append(int(np.exp(power)))
```

```python
plt.plot(np.log(n0_list),np.log(c0),label="c=0.0")
plt.plot(np.log(n0_list),np.log(c01),label="c=0.1")
plt.plot(np.log(n0_list),np.log(c02),label="c=0.2")
plt.plot(np.log(n0_list),np.log(c03),label="c=0.33")
plt.plot(np.log(n0_list),np.log(c04),label="c=0.4")
plt.plot(np.log(n0_list),np.log(c05),label="c=0.5")
#plt.plot(np.log(n0_list),(-1/6)*np.log(n0_list)-1,label="Line with slope -1/6")
#plt.plot(np.log(n0_list),np.log(c_half),label="c=1/2")
plt.title("TV Error varying c")
plt.ylabel("Log of the TV error")
plt.xlabel("Log of the number of samples n0")
plt.legend()
plt.grid()
plt.savefig(f"WingIT+PI_TV_error_n0=d_Varyc.pdf")
```

```python
plt.plot(np.log(n0_list),np.log(PIc0),label="c=0.0")
plt.plot(np.log(n0_list),np.log(PIc01),label="c=0.1")
plt.plot(np.log(n0_list),(-1/6)*np.log(n0_list)+0.7,label="Line with slope -1/6")
#plt.plot(np.log(n0_list),np.log(c_half),label="c=1/2")
plt.title("PI TV Error varying c")
plt.ylabel("Log of the TV error")
plt.xlabel("Log of the number of samples n0")
plt.legend()
plt.grid()
plt.savefig(f"PI_TV_error_n0=d_Varyc.pdf")
```
